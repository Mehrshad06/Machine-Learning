```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'a': [1, 2, 3, 4],
    'b': [5, 6, 7, 8],
    'c': ['red', 'blue', 'green', 'yellow'],
    'd': [10, 40, 30, 20] 
})

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'a': np.random.rand(50),
    'b': np.random.rand(50) * 10,
    'c': np.random.choice(['red', 'blue', 'green'], 50)
})

data['d'] = np.random.rand(50) * 100
data['d'] = np.abs(data['d']) * 100

print(data.head())
```
